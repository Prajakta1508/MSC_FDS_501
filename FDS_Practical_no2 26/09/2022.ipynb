```python
import matplotlib.pyplot as plt
plt.figure(figsize=[15,5])
countries=['USA','Brazil','Russia','Spain', 'UK','India'],
totalCases=(2026493,710887,476658,288797,287399,265928)
totalDeaths=(113055,37312,5971,27136,40597,7473)
for i in range(len(countries)):
  plt.bar(countries[i],totalCases[i],bottom=totalCases[i] - totalDeaths[i],color='black')
  plt.bar(countries[i],totalCases[i] - totalDeaths[i],color='red')

  plt.legend(labels= ['Total Deaths','Total Cases'])
  plt.title("bar plot")
  plt.xlabel('countries')
  plt.ylabel('Cases')
  plt.savefig('3barplot.png')
  plt.show()
  
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=[15,10])
totalDeaths=[113055,37312,5971,7473,33964]
totalRecovery=[773480,325602,230688,129095,166584]
activeCases=[1139958,347973,23999,29360,34730]
country=['USA','Brazil', 'Russia','India','Italy']
x=np.arange(len(totalDeaths))
plt.bar(x,totalDeaths,color ='Black',width=0.25)
plt.bar(x + 0.25,totalRecovery,color ='g',width=0.25)
plt.bar(x + 0.25,activeCases,color ='b',width=0.25)
plt.legend(['totalDeaths','totalRecovery','activeCases'])
plt.xticks([i+0.25 for i in range (5)],country)

plt.title("Bar plot represnting the total  deaths,total recovered,cases and active cases country wise")

plt.xlabel('Countries')
plt.ylabel('Cases')
plt.savefig('4BarPlot.png')

plt.show()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt

Data={'Tasks': [300,500,700],
      'Task Type' :['Tasks Pending','Tasks Ongoing','Tasks Complete']
}
df= DataFrame(Data)
df.set_index('Task Type',inplace=True)
df.plot.pie(y='Tasks', figsize=(10,10),autopct='%1.1f%%', startangle=90)

```

```python
import numpy as np
import matplotlib.pyplot as plt
labels=['civil','Electrical','Mechanical','Chemical']
sizes=[15,50,45,10]

explode=(0.1,0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,
       explode=explode,
       labels=labels,
       autopct='$1.1f%%',
       shadow=True,
       startangle=90)
ax.axis('equal')
ax.set_title('engineering Diciplines')
plt.show()
```
